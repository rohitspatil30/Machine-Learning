```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
print("Imported!")
```

```python
# collectiong data
file = pd.read_csv("./heart-missing-classification-dataset.csv")
file.isna().sum()
```

```python
file.head(15)
```

```python
print(file.info())
```

```python
x = file.drop("target",axis = 1)
y = file["target"]

x
```

```python
# fillna on x or features

x = x.fillna(x.mean())
x.isna().sum()
```

```python
y
```

```python
import random
y = y.fillna(random.randint(0,1))
y
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(3)

x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.2)

x_train
```

```python
from sklearn import tree

model1 = tree.DecisionTreeClassifier()

model1.fit(x_train,y_train)
```

```python
print(f"The accuracy of the Decision tree classifier model is :{model1.score(x_test,y_test)*100}%")
```

```python
from sklearn import neighbors

model2 = neighbors.KNeighborsClassifier()

model2.fit(x_train,y_train)
```

```python
print(f"The accuracy of the K nearest Neighbors classifier model is :{model2.score(x_test,y_test)*100}%")
```

```python
from sklearn import naive_bayes

model3 = naive_bayes.BernoulliNB()

model3.fit(x_train,y_train)
```

```python
print(f"The accuracy of the K nearest Neighbors classifier model is :{model3.score(x_test,y_test)*100}%")
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
X, y = make_classification(n_features=13, random_state=0)
clf = make_pipeline(StandardScaler(),
                    LinearSVC(dual="auto", random_state=0, tol=1e-5))
clf.fit(X, y)
```

```python
clf.score(x_test,y_test)
```
