```python
# we know that we have missing values, also we know that we have all numbers in the dataset
# removing missing values
# imputing -> filling the missing values by a number.

file = file.fillna(file.mean()) #--> filling the file data 
```

```python
# training 
# vertical split -> splitting the data into features(input-> what should 
# you learn to predict the output) and labels(output-> what you want to predict)

x = file.drop("Final",axis=1)
y = file["Final"]
x
```
