# Scikit Learn 

## What is Scikit learn?
answer: Scikit learn is a python based library for machine learning. It is used to build the machine learning model on top of Data. Also it is used to evaluate the model on the basis of predictions it will make.

## Why SKLearn?
1. Sklearn is built on the top on numpy and matplotlib(python also).
2. Has many Built-in Models that can be directly used to build models.
3. Have many methods to evaluate the models.
4. It has very well-directed APIs.

### Steps for the workflow:
1. An end-to-end workflow including sklearn imports
2. Getting the Machine Learning Data ready.
3. Choosing the Model, according to the data and the problem statement.
4. Fitting a model or training the model using the data.
5. Making the predictions with data, using testing data.
6. Evaluating model performances.
7. Improving the predictions of the model.
8. Saving the model.
9. Reloading the re-training.
10. Transfer Learning.

```python
#importing the libraries.
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn as sk
import numpy as np 
import pandas as pd
print("Imported!")
```

```python
file = pd.read_csv("../datasets/heart.csv")
file
```

**Now that we have the data it is important to define the data**

First thing everything in the data is either a feature or a label.
feature is the input or the things that we know about the data, they help us describe the data in such a way that finding accurate predictions becomes simpler.
Labels are the outputs we expect from the machine.

Initially we give the features and the expected labels both to the machine, so that the machine could learn what type of data produces what kind of label.
This is important step as when it will be tested on the new esting data, it can give expected labels.

Now, Here first step is to seperate the features from the labels and mark/store them into different variables called as the x for the features and y for the labels.



```python
x = file.drop("target", axis =1) #if we drop/delete the target label column then the rest are the features for training.
y = file["target"] # the column or series is the target or the label.

# let's print the type of data.
print(type(y))
print(y.dtype) # type of data type in the column/series

# same with dataframe
print(type(x))
x.info()
```

```python
# getting the model, selecting the right model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
# 3 fit the model to training data..
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)#here the data will be split in 80:20 ratio from both x,y varaibele .
```

```python
clf.fit(x_train,y_train);
```

```python
y_label= clf.predict(x_train)
y_label
```

```python
x_test
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
#4 . evaluate

clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
#imrove the model
#try different n_estimators 
np.random.seed(2)
for i in range(10,100,10):
    print(f"Trying model with {i} extimators...")
    clf =RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"model accuracy on the test set: {clf.score(x_test,y_test)*100:.2f}")
    print(" ")
```

```python
# Save the model and load it.
import pickle
pickle.dump(clf, open("random_forest.pkl","wb"))
print("Saved!!")
```

```python
#load the model again
load_model = pickle.load(open("random_forest.pkl","rb"))
load_model.score(x_test,y_test)
```
