```python
# standard imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn

print("Imported!")
```

```python
# loading standard boston housiing data set
# inhouse datasets of sklearn
'''
breast cancer
diabetes
wine


'''
from sklearn.datasets import load_diabetes
f = load_diabetes()
# file the data is always imported as dictionary
# convert the python dictionary to pandas dataframes
file = pd.DataFrame(f["data"],columns=f["feature_names"])

# creating the target column
file["target"] = pd.Series(f["target"])

file
```

```python
from sklearn.linear_model import Ridge

# setup random seed
np.random.seed(42)

# vertical
x = file.drop("target",axis =1)
y = file["target"]

# vertical split
```

```python
# training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# machine learning

model = Ridge()

model.fit(x_train,y_train)
model.score(x_test,y_test)
```

 # how do we improve this score?
 

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```
