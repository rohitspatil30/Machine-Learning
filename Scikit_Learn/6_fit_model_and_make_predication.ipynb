## Some theory regarding the model fitting:

* `model.fit(x_train,y_train)` is responsive for training of the data.Here in this we give the `features` as the input and `labels` too as the training input, the alorithm learns as the patterns inside  the input fetures and its relation with the labels.
* `model.score(x_test,y_test)` on the other hand gives the testing data for the prediction compares it with y_test and the gives the accuracy score.

```python
# predict proba returns probabilities of the classification - 
etc.predict_proba(x_test[:5])
```

now here look at the first 5 predticions and the first 5 probablities





if the 0th index probablity is greater than 50 then the output is 0 else the index 1 is having the greater probablity then the output is 1. here first 4 are 1 coz their probablitites are greater than 0.50.

note: that the summation of both the probablitites for 0 and 1 adds upto 1.0  which is called as total probability

## It's application

now imagine this probability as the confidence 34% percent confidence that the patient is having confidence can be written as the 0.34, but is this confidence enough for actual model, no , not even 70% of the probablity is good to say that the person is having heart disease so in that case, we always go for if condition `if(etc.predict_proba[:5][1] > 0.85): print("hear disease")`

```python
# here is small heart disease predictor code which uses the confidence or the probablity of the x_test to get the heart disease occurance.

tup = etc.predict_proba(x_test[:20])
for i in range(0,len(tup)):
    if(tup[i][1]>0.75):
        print("The Patient is likely to have heart disease.because the percentage of it is: ",tup[i][1])
    elif(tup[i][0]<0.50):
        print("The patient is less likely to have a heart disease, because the percentge of not having it is: ",tup[i][0])
    
```
