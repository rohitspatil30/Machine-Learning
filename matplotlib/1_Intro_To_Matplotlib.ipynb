# Difference between the pyplot api and object oriented api.
**Prefer the object orientd API over Pyplot API** as it allows the better maintainace of the figures and the axes. have better flexiblity and control over the data.
There are *Three diiferent ways* to plot the data using the object oriented API.



```python
# 1st method 
figure = plt.figure() # createsa white background
axis = figure.add_subplot()# adds the axis to it.
axis.plot(study_time, marks_obtained)
plt.show()
```

```python
# 2nd method
figure = plt.figure() #creates the white background
axis = figure.add_axes([1,1,1,1],"Rohit") #tuple ( left, bottom, width, height)
axis.plot(study_time,marks_obtained)
plt.show()
```

```python
# 3rd method (recommended)
figure, axis = plt.subplots()
axis.plot(study_time,marks_obtained)
plt.show()
```

# Step-wise workflow of matplotlib

```python
#step 0 : Import matplotlib.pyplot
%matplotlib inline
import matplotlib.pyplot as plt

# step 1: organize the data
print(study_time, marks_obtained,"\n\n")

# step 2: Setup the plot
figure,axis = plt.subplots(figsize=(7,7))

#step 3: plot the data points on the figure
axis.plot(study_time, marks_obtained);

#step 4: Customize  the plots and labels
axis.set(title="Study Marks Ratio", xlabel = "Study time", ylabel = "Marks obtained");

#step 5: Save the grapdh and share it.
figure.savefig("../graphs_and_images/study-marks-ratio.png")



```
