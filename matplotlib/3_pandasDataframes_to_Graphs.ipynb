# plotting from pandas Dataframes
The actual data is not plotted directly using the numpy array but using the dataframes or tables , it is tables which is converted to the visual representation for better understanding.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print("Imported!")
```

```python
file = pd.read_csv("../pandas/class-grades.csv")
print(file)
```

```python
file = file.fillna(file.mean())
file #filled the missing values using the mean of each column
```

```python
#plotting the graph
file.hist("Tutorial"); 
#directly specify the columns name in the hist value as the function takes the series or list as input and every column is
# list in itself.
```

```python
figure, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10));
ax1.scatter(file["Roll no"],file["Assignment"]);
ax1.set(title="Assignment marks",xlabel="Roll numbers",ylabel="Assignment marks")
ax2.bar(file["Roll no"],file["Tutorial"]);
ax2.set(title="Tutorial marks",xlabel="Roll numbers",ylabel="Tutorial marks");
ax3.bar(file["Roll no"],file["Final"]);
ax3.set(title="Final marks",xlabel="Roll numbers",ylabel="Final marks");
ax4.bar(file["Roll no"],file["Midterm"]);
ax4.set(title="Midterm marks",xlabel="Roll numbers",ylabel="Midterm");


```

This does not look like ideal data visualisation.



```python
#pyplot method to produce a quick graph
over_50_marks = file[file["Final"]>=50]
over_50_marks.plot(kind="bar",x="Roll no",y="Final",title="Final marks");
```

```python
over_50_marks.plot(kind="scatter",x="Roll no",y="Final",title="Final marks",c="Final");
```

```python
#oo method
# two methods by calling the pylot dataframe to plot the table and the second method is to call the axis to make x and y of the file series.
fig,ax = plt.subplots(figsize=(10,10));
# call the dataframe
over_50_marks.plot(kind="scatter",
         x="Roll no",
         y="Tutorial",
         c="Tutorial",
            ax=ax);
```

```python
fig,ax = plt.subplots(figsize=(10,10));
ax.scatter(x=over_50_marks["Roll no"],y=over_50_marks["Final"],c=over_50_marks["Final"]);
ax.set(title="Scatter plot using oo method");
ax.axhline(over_50_marks["Final"].mean(), ls="--");
```

# Shared values on same x or y axis on a subplot

```python
fig, ((ax1),(ax2),(ax3),(ax4)) = plt.subplots(nrows=4,ncols=1,figsize=(10,30),sharex=True)

# graph ax1
scatter = ax1.scatter(x=over_50_marks["Roll no"],y=over_50_marks["Tutorial"],c=over_50_marks["Tutorial"]);
ax1.legend(*scatter.legend_elements(), title="Legends")
ax1.set(ylabel="Tutorial",title="Tutorial graph");
ax1.axhline(over_50_marks["Tutorial"].mean(),ls="--");
ax1.set_xlim([1,85]);
ax1.set_ylim([50,100])

# graph ax2
ax1.scatter(x=over_50_marks["Roll no"],y=over_50_marks["Assignment"],c=over_50_marks["Assignment"]);
ax1.set(ylabel="Assignment",title="Assignment graph");
ax1.axhline(over_50_marks["Assignment"].mean());

# graph ax3
ax2.scatter(x=over_50_marks["Roll no"],y=over_50_marks["Assignment"],c=over_50_marks["Assignment"]);
ax2.set(ylabel="Assignment",title="Assignment graph");
ax2.axhline(over_50_marks["Assignment"].mean(),ls="--");


# graph ax4
ax3.scatter(x=over_50_marks["Roll no"],y=over_50_marks["Midterm"]);
ax3.set(xlabel="Roll no",ylabel="Midterm",title="Midterm graph");
ax3.axhline(over_50_marks["Midterm"].mean(),ls="--");

# graph ax5
ax4.scatter(x=over_50_marks["Roll no"],y=over_50_marks["Final"]);
ax4.set(xlabel="Roll no",ylabel="Final",title="Final graph");
ax4.axhline(over_50_marks["Final"].mean(),ls="--");

fig.suptitle("Class Grades\nThe Final Grades for the annual year",fontweight="bold",fontsize="24");

#adding grid to the plots
plt.style.available;
```
