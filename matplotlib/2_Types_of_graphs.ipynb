```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
print("Imported!")
```

# Creating a line plot


```python
#setting up the data
basic_array_x= np.array([3,14,25,28,45,50])
basic_array_y =np.array([10,46,40,49,43,50])

#setting up the plot
figure,axis = plt.subplots(figsize=(7,7))

# putting the datapoints 
axis.plot(basic_array_x,basic_array_y);

# cutomize the plot
axis.set(title="Line Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");

```

# Creating a scatter plot
A scatter plot is a type of graph that shows the relationship between two variables in a data set. The variables are plotted on a two-dimensional plane or Cartesian system. The independent variable is plotted on the X-axis, and the dependent variable is plotted on the Y-axis. Each data point is represented by a dot. The position of each dot on the horizontal and vertical axis indicates the values for an individual data point.
Scatter plots are used to observe relationships between variables. They are a useful diagnostic tool for determining association. However, if an association exists, the plot may or may not suggest an underlying cause-and-effect mechanism
.

```python
#scatter plot 
figure,axis = plt.subplots()
axis.scatter(basic_array_x,basic_array_y)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");
```

## Scatter plot for exponential values

```python
#scatter plot on a big random data
import math
a1 = list(range(2,25)) #x should be increasing only else it will go back and forth
a2 = [math.exp(num) for num in a1 ]
figure,axis = plt.subplots() 
axis.scatter(a1,a2)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");

a1 = list(range(2,25)) #x should be increasing only else it will go back and forth
a2 = [math.log(num) for num in a1 ]
figure,axis = plt.subplots() 
axis.scatter(a1,a2)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");


figure,axis = plt.subplots() 
axis.scatter(a1,np.sin(a1))
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");

figure,axis = plt.subplots()
axis.plot(a1,a2)
axis.set(title="Line Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");
```

```python
#scatter plot on a big random data
a1 = list(range(2,102)) #x should be increasing only else it will go back and forth
a2 = np.random.randint(2,100,size=(100)) # y can vary more or less
figure,axis = plt.subplots() 
axis.scatter(a1,a2)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");

figure,axis = plt.subplots()
axis.plot(a1,a2)
axis.set(title="Line Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");
```

```python
#horizontal bar
a1 = list(range(2,25)) #x should be increasing only else it will go back and forth
a2 = [math.pow(num,2) for num in a1 ]
figure,axis = plt.subplots()  #creates a new canvas each time
axis.bar(a1,a2)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");
figure,axis = plt.subplots() 
axis.barh(a1,a2)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");
figure,axis = plt.subplots() 
axis.scatter(a1,a2)
axis.set(title="Scatter Plot",
        xlabel= "X-axis data",
        ylabel= "y-axis data");
```

# Subplots

Subplots allow you to make more than 1 plot in one entire figure so that you can have multiple graphs in one canvas screen.


```python
#syntax: for figure define the tuple graphs inside a tuple, (ax1,ax2)-> 2 graphs pack * 3  -> then specify the space you need in nrows
#  and ncols and then ax1
figure, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows = 2,ncols=2, figsize = (10,10))

# filling data in theses values
ax1.plot(basic_array_x,basic_array_y);
ax2.scatter(a1,np.exp(a1));
ax3.bar(dict.keys(),dict.values());
ax4.hist(np.random.randint(1,1000,size=(1000))); #randn stands for normal distribution value for 1000
```

# You know sometimes you just don't want to make variable or tuples of each graph so, just access it by index.

# original syntax:
figure, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows =2,ncols =2, figsize = (10,10));

here we have  2 tuples inside a tuple. each tuple here represents the element of the plot. 



```python
#new syntax:
figure , axis = plt.subplots(nrows=2,ncols=2,figsize=(10,10))#just define the number of rows and cols and not the tuples
# rather access it using index.
axis[0,0].plot(basic_array_x,basic_array_y);
axis[0,1].scatter(a1,np.exp(a1));
axis[1,0].bar(dict.keys(),dict.values());
axis[1,1].hist(np.random.randint(1,1000,size=(1000))); #randn stands for normal distribution value for 1000                        
```
